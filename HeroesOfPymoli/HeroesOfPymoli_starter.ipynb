```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
pd.set_option('mode.chained_assignment', None)
# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
total_player = len(purchase_data_df["SN"].unique())
total_player_df = pd.DataFrame({"Total Players":[total_player]})
                                                 
total_player_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python

unique_items = len(purchase_data_df["Item Name"].unique())
average_price = purchase_data_df["Price"].mean()
total_purchases = len(purchase_data_df)
total_revenue = purchase_data_df["Price"].sum()
```

```python
analysis_totals_df = pd.DataFrame({"Total Unique Items":[unique_items],
                                   "Average Price":[average_price], 
                                   "Total Purchases": [total_purchases],
                                   "Total Revenue":[total_revenue]})
analysis_totals_df["Average Price"]=analysis_totals_df["Average Price"].map("${:.2f}".format)
analysis_totals_df["Total Revenue"] = analysis_totals_df["Total Revenue"].map("${:.2f}".format)
analysis_totals_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#make a database of personal data, filtering out repeated names
unique_df = purchase_data_df[["SN", "Age", "Gender"]].drop_duplicates()
gender_demo_df = unique_df.groupby(["Gender"])
counted_gender_df = gender_demo_df.count()
counted_gender_df = counted_gender_df.rename(columns={"SN": "Total"})
del counted_gender_df["Age"]
counted_gender_df["Percentage"] = (counted_gender_df["Total"]*100/counted_gender_df["Total"].sum())

counted_gender_df["Percentage"]=counted_gender_df["Percentage"].map("{:.2f}%".format)
counted_gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#group original df by gender
gender_totals_df = purchase_data_df.groupby(["Gender"])

#acquire variables of count, mean and sum
gendered_purchase_df = gender_totals_df.count()
mean_gender_price = gender_totals_df.mean()
summed_value_df = gender_totals_df.sum()

#slim those down into a single column each

purchase_count = gendered_purchase_df["Purchase ID"]
gendered_average = mean_gender_price["Price"]
total_gen_purchases = summed_value_df["Price"] 

#create a dataframe that can hold each new column
total_purchase_df = gender_totals_df.count() 
total_purchase_df["Purchase Count"] = gendered_purchase_df[["Purchase ID"]]
average_gender_list = mean_gender_price.loc[:,"Price"]
total_purchase_df["Average Price"] = average_gender_list
total_purchase_df["Total Purchase Value"] = total_gen_purchases

#slim new dataframe down to just relevant columns
incomplete_df = total_purchase_df[["Purchase Count", "Average Price","Total Purchase Value"]]
average_total = incomplete_df["Total Purchase Value"]/counted_gender_df["Total"]
#create one new column to find total purchase per person
incomplete_df["Average Total Purchase per Person"] = average_total

#reformat to mark currencies
incomplete_df["Average Total Purchase per Person"]=incomplete_df["Average Total Purchase per Person"].map("${:.2f}".format)
incomplete_df["Average Price"] = incomplete_df["Average Price"].map("${:.2f}".format)
incomplete_df["Total Purchase Value"] = incomplete_df["Total Purchase Value"].map("${:.2f}".format)
incomplete_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

#make bins for the ages
bins=[0, 9, 14, 19, 24, 29, 34, 39, 200]
bin_labels=["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39", "40+"]

#attach bins to DF
age_bins = pd.cut(unique_df["Age"], bins, labels=bin_labels, include_lowest=True)
unique_df["Age Range"] = age_bins

#group by bins
grouped_range=unique_df.groupby(["Age Range"])
counted_unique = grouped_range.count()

#make new dataframe to hold desired info
age_demos_df = counted_unique[["Age"]]
age_demos_df["Percentage"] = (age_demos_df["Age"]*100/total_player)
age_demos_df = age_demos_df.rename(columns={"Age":"Total Count"})

age_demos_df["Percentage"]=age_demos_df["Percentage"].map("{:.2f}%".format)
age_demos_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins=[0, 9, 14, 19, 24, 29, 34, 39, 200]
bin_labels=["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39", "40+"]
age_bins = pd.cut(purchase_data_df["Age"], bins, labels=bin_labels, include_lowest=True)
purchase_data_df["Age Range"] = age_bins
aged_groups = purchase_data_df.groupby(["Age Range"])
aged_counts = aged_groups.count()
aged_counts = aged_counts.rename(columns={"Purchase ID":"Purchase Count"})
holder_df = aged_counts[["Purchase Count"]]

mean_aged_price = aged_groups.mean()
summed_aged_df = aged_groups.sum()
holder_df["Average Purchase Price"] = mean_aged_price["Price"]
holder_df["Total Purchase Value"] = summed_aged_df["Price"]
holder_df["Avg Total Purchase per Person"] = holder_df["Total Purchase Value"]/age_demos_df["Total Count"]

holder_df["Avg Total Purchase per Person"]=holder_df["Avg Total Purchase per Person"].map("${:.2f}".format)
holder_df["Average Purchase Price"] = holder_df["Average Purchase Price"].map("${:.2f}".format)
holder_df["Total Purchase Value"] = holder_df["Total Purchase Value"].map("${:.2f}".format)
holder_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_groups = purchase_data_df.groupby(["SN"])
sn_count = sn_groups.count()
sn_count = sn_count.rename(columns={"Purchase ID":"Purchase Count"})
#make DF to hold info
spending_df = sn_count[["Purchase Count"]]
#create columns with required information
sn_mean = sn_groups.mean()
sn_sum = sn_groups.sum()
spending_df["Average Purchase Price"] = sn_mean["Price"]
spending_df["Total Purchase Value"] = sn_sum["Price"]
sorted_df = spending_df.sort_values( "Total Purchase Value", ascending=False)

sorted_df["Average Purchase Price"] = sorted_df["Average Purchase Price"].map("${:.2f}".format)
sorted_df["Total Purchase Value"] = sorted_df["Total Purchase Value"].map("${:.2f}".format)
sorted_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
grouped_items = items_df.groupby(["Item ID", "Item Name"])
items_counted_df = grouped_items.count()
items_counted_df= items_counted_df.rename(columns={"Price":"Purchase Count"})
popularity_df = items_counted_df[["Purchase Count"]]

items_sum = grouped_items.sum()
items_mean= grouped_items.mean()

popularity_df["Item Price"] = items_mean["Price"]
popularity_df ["Total Purchase Value"] = items_sum["Price"]

popularity_format_df=popularity_df
#sorting for profit before formatting goes into effect
profit_df = popularity_df.sort_values( ["Total Purchase Value"] , ascending = False)
#formatting
popularity_format_df["Item Price"] = popularity_format_df["Item Price"].map("${:.2f}".format)
popularity_format_df["Total Purchase Value"] = popularity_format_df["Total Purchase Value"].map("${:.2f}".format)
popularity_sort_df = popularity_format_df.sort_values( "Purchase Count", ascending=False)
popularity_sort_df.head()
```
